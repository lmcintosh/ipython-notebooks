```python
import numpy as np
import matplotlib

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.plot(np.exp(-np.linspace(0,10,2000)),'k',linewidth=4)
```

```python
fig = plt.gcf()
fig.set_size_inches((10,6))

alpha = 100
theta = 0
plt.plot(np.linspace(-1,1,2000),1./(1+np.exp(-alpha*(np.linspace(-1,1,2000) - theta))),'k',linewidth=10)
# plt.vlines(0,-0.2,1.2,colors='k',linestyle='--',linewidth=1.5)
plt.ylim(-0.02,1.02)
plt.xlim(-.2,.2)
plt.axis('off')
```

```python
p = np.linspace(0,1,1000)
n = np.arange(10)

var = p*(1-p)

multi_var = []
means = []
for ni in n:
    for pi in p:
        multi_var.append(ni*pi*(1-pi))
        means.append(ni + pi)
```

```python
plt.plot(p, p*(1-p))
```

```python
from aesthetics.plotting import adjust_spines
```

```python
fig = plt.gcf()
fig.set_size_inches((4,3))

plt.plot(means, multi_var, 'k', linewidth=8)
plt.plot(means, means, 'r', linewidth=8)
adjust_spines(plt.gca())
plt.ylabel('Variance', fontsize=18)
plt.xlabel('Mean', fontsize=18)
```
